Задание №1. 
Когда используется критерий Стьюдента, а когда Z –критерий?

Критерий Стьюдента применяется для проверки равенства средних значений в двух выборках. Т.е. сравниваются две полученные выборки. 
t-тест приемлем, когда размеры выборки малы, n < 30.

Критерий Стьюдента используется для сравнения средних значений двух групп. Также его можно использовать для сравнения средних значений двух выборок, если данные для двух групп распределены нормально.
Закон Стьюдента позволяет оценить то, насколько средние значения двух групп отличаются.


Z-критерий используется для сравнения наблюдаемой выборки с теоретической. Т.е. одна полученная (имеющаяся) выборка сравнивается с теоретичекой (заданной) величиной. 
Z-тест используется, когда размер выборки большой, т.е. n > 30

Закон Z-критерия используется для проверки гипотезы о равенстве среднего значения генеральной совокупности с известным значением.
Данный критерий используется для проверки гипотезы о нормальности распределения данных в выборке с известным значением среднего значения.Это позволяет оценить точность среднего значения измерения в генеральной совокупности, а также понять, насколько данные в выборке представляют собой нормальное распределение.

t-тест часто понимается как статистический тест, который используется для сопоставления и анализа того, отличаются ли средства двух популяций друг от друга или нет, если отклонение в качестве неизвестно. В отличие от этого Z-тест может представлять собой параметрический тест, который применяется, когда известно о качественном отклонении, для выяснения того, отличаются ли средства двух наборов данных друг от друга.
Одним из важных условий принятия t-теста является то, что дисперсия популяции неизвестна. И наоборот, дисперсия популяции должна быть известна или предположительно известна только в случае z-теста.

Задача №2. 
Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.

H0: M = 17 (нулевая гипотеза)

H1: M = 17,5 (не равен 17 - альтернативная гипотеза)

```python
U_fact = (17.5 - 17) / np.sqrt(4) * np.sqrt(100)
```

```python
U_fact
```

```python
U_cr = 1.645
```

Ответ: т.к. U_fact (2,5) > U_cr (1,645) отклоняем нулевую гипотезу и принимаем альтернативную.
Средний диаметр шариков для подшипников, изготовленных автоматическим станком, не равен 17 мм.

Задача №3.
Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? (Провести двусторонний тест.)

H0: M = 200

H1: M = 200

```python
X = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
mean_X = X.mean()
std_X = X.std(ddof=1)
```

```python
t_fact = (mean_X - 200) / std_X * np.sqrt(10)
t_fact
```

```python
t_cr = 3.25
```

т.к. модуль t_fact (1,0651) < t_cr (3,25) нулевая гипотеза принимается.

Задача №4.
Есть ли статистически значимые различия в росте дочерей?
Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165
Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160.


```python
mothers = [172, 177, 158, 170, 178, 175, 164, 160, 169, 165]
daughters = [173, 175, 162, 174, 175, 168, 155, 170, 160, 163]

stats.ttest_rel(mothers, daughters)
```

Ответ:
Так как p-value = 0,5228 > 0,05 (a)
H0 не отвергаем.  

Статистически значимых различий нет.
